# Sending request to server ( Web-stie)

If the response 403 Or forbidden the enter the headers


```python
import requests

# Define the headers
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

# Send a GET request to the URL with the specified headers
url = 'https://www.themoviedb.org/leaderboard'
response = requests.get(url, headers=headers)

# Get the content of the response
web_page = response.text

# Print or process the web page content
print(web_page)
```

# Converting it into an html site 

```python
parser = BeautifulSoup(web_page,'lxml')
```

```python
print(parser.prettify())
```

# Extracting the heading to check program work or not
